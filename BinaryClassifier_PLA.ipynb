```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
#linearly separable data at random
X1=np.random.randint(low=0,high=100,size=10)
x2=np.random.randint(low=10,high=40,size=5)
x3=np.random.randint(low=70,high=100,size=5)
```

```python
x1=np.array(X1)
x2=np.array(x2)
X2=np.concatenate((x2,x3))
```

```python
#dataset of 20 random linearly separable points
X=np.array([X1,X2])
X=X.T
print(X)

#augmenting X to include scalar constant
one=np.array([1 for i in range(len(X1))])
aug_X=np.array([one,X1,X2])
aug_X=aug_X.T

print(aug_X)
```

```python
#result of classification
Y1=np.array([-1 for i in range(len(x2))])
Y2=np.array([1 for i in range(len(x3))])
Y=np.concatenate((Y1,Y2))
print(Y)
```

```python
#plotting the data
for i in range(len(X1)):
    if(Y[i]==-1):
        plt.scatter(X[i,0],X[i,1],color='red')
    else:
        plt.scatter(X[i,0],X[i,1],color='blue')
```

```python
#generating random w according to PLA
w=np.random.randint(low=0,high=100,size=3)
w=np.array(w)
print(w)
```

```python
#PERCEPTRON LEARNING ALGORITHM
count=1
while(1):
    li=[]
    print("Epoch: ",count)
    for i in range(len(X1)):
        if(np.dot(aug_X[i],w)>0 and Y[i]>0):
            li.append(1)
        elif((np.dot(aug_X[i],w)<0 and Y[i]>0) or (np.dot(aug_X[i],w)>0 and Y[i]<0)):
            li.append(0)
            w=w+np.dot(Y[i],aug_X[i])
            print('updated w:',w)
        else:
            li.append(-1)
    
    li=np.array(li)
    if(li.all()==Y.all()):
        print("Last epoch: ",count+1)
        break
    else:
        count+=1

print("Final w:",w)
```

```python
#finding points to plot for linear decision boundary
random_x=np.random.randint(low=0,high=100,size=100)
random_y= -(w[0]/w[2]) - (w[1]/w[2])*random_x
```

```python
#plotting line on the graph
plt.plot(random_x,random_y,color='green',label='Linear Decision Boundary')

for i in range(len(X1)):
    if(Y[i]==-1):
        plt.scatter(X[i,0],X[i,1],color='red')
    else:
        plt.scatter(X[i,0],X[i,1],color='blue')

#labelling the plot
plt.title("Perceptron Learning Algorithm")
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```
